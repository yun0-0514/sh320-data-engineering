````python
import requests
from bs4 import BeautifulSoup
````

### <strong> 예제 1: 메타버스 키워드 검색 및 뉴스 타이틀 추출
----

````python
# 웹페이지 주소
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EB%A9%94%ED%83%80%EB%B2%84%EC%8A%A4'

# [+] HTTP GET Request로 HTML 문서 받아오기(get은 서버로 부터 원하는 정보를 요청하는 것,post는 서버에 정보를 보내는 것)
# HTTP GET Request로 HTML 문서 받아오기
res = requests.get(url)
res
````

````python
# [+] HTML 문서로부터 텍스트 데이터 덩어리 추출
html_doc = res.text
html_doc
````

#### <strong> BeautifulSoup 파서 사용하기</strong>
+ ```html.parser```: HTML 문서 전용 파서
+ ```xml```: XML 문서 전용 파서

````python
# [+] Beautiful Soup에서 제공하는 HTML 파서 생성
soup = BeautifulSoup(html_doc, 'html.parser')#여러가지 parser가 있지만 html.parser를 사용
````

```BeautifulSoup.prettify()```: 문서 정보를 시각적으로 보기 좋게 재구조화

````python
print(soup.prettify())#prettify()는 HTML 문서의 구조를 보기 좋게 정리해주는 함수
````

`BeautifulSoup.find_all()`: 특정 태그 정보들을 검색

````python
# [+] 문서 내에 모든 <a> 태그들을 검색
soup.find_all('a')
````

````python
# [+]뉴스에 해당되는 태그들만 검색하기
# "class" 속성이 "news_tit"인 <a> 태그
news = soup.find_all('a',{'class':'news_tit'})#검색 조건은 딕셔너리 형태로 제공 
news
````

````python
# 첫 번째 뉴스//특정 정보에 대한 속성(ㅇattribute)를 가져오기 위해 get()사용//
news[0]
````

````python
# [+] 뉴스의 URL 정보 가져오기
for news_obj in news:
    print(news_obj.get('href'))
````

````python
# [+] 뉴스의 제목 정보 가져오기
for title in news:
    print(title.get('title'))
````

````python
# [+] 모든 뉴스의 제목 출력하기
for news_obj in news:
    print(title.get('title'))
````
